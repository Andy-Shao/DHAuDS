# HuBERT running example
## HuBERT BASE

```python
import torchaudio

bundle = torchaudio.pipelines.HUBERT_BASE
hubert_base = bundle.get_model()
print(hubert_base)
```

```python
import torch 

device = 'cuda' if torch.cuda.is_available() else 'cpu'
hubert_base = hubert_base.to(device=device)

waveform = torch.ones(2, 48000*60)
waveform = torchaudio.functional.resample(waveform, orig_freq=48000, new_freq=bundle.sample_rate)
waveform = waveform.to(device)
features, _ = hubert_base.extract_features(waveform)
for f in features:
    print(f.shape)
print(f'totoal number of feature is:{len(features)}')
```
