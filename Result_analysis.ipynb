# Impact of Momentum Analysis
## AMAuT

```python
# AMAuT RS ENQ L1
import pandas as pd

iom = pd.read_csv('./result/ReefSet/AMAuT/Analysis/IoM_analysis.csv', header=0, delimiter=',')
iom.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(iom['Step'][1:], iom['AuT-RS-HM'][1:], label='HM', marker='o')
plt.plot(iom['Step'][1:], iom['AuT-RS-LM'][1:], label='LM', marker='o')
plt.xlabel('Epoch', fontsize=18)
plt.ylabel('ROC-AUC', fontsize=18)
# plt.ylim(ymin=.76)
# plt.title('Comparison ROC-AUC between high momentum (HM) and low momentum (LM) when performing AMAuT ENQ-L1 on RS')
plt.legend()
plt.grid(True)
# plt.savefig(
#     'IoM_AuT-RS_ENQ-L1.png',
#     format='png',
#     dpi=300,
#     bbox_inches='tight',
#     facecolor='white',
#     edgecolor='none'
# )
plt.show()
```
