```python
import torch 
from transformers import ASTModel, AutoFeatureExtractor

ast = ASTModel.from_pretrained("MIT/ast-finetuned-audioset-10-10-0.4593")
feature_extractor = AutoFeatureExtractor.from_pretrained("MIT/ast-finetuned-audioset-10-10-0.4593")

# inputs = torch.ones(1, 128, 1024)
# outputs = ast(inputs).pooler_output
# print(outputs.shape)

# dummy waveform: 1 second of silence
waveforms = [torch.randn(16000) for _ in range(4)]
inputs = feature_extractor([w.numpy() for w in waveforms], sampling_rate=16000, return_tensors="pt")['input_values']

outputs = ast(inputs)

print(outputs.last_hidden_state.shape)  # (1, num_patches, hidden_size)
print(outputs.pooler_output.shape)      # (1, hidden_size)
print(inputs.shape)
```
