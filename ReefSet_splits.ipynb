Since ReefSet is without an official splits between traning and testing. This script takes the responsibility for generating the ReefSet splits. We splits the ReefSet as 7:3 between training and testing sets among each labels.

```python
import os
dataset_root_path = '/root/data/ReefSet_v1.0' 

import pandas as pd

data_meta = pd.read_json(os.path.join(dataset_root_path, 'reefset_annotations.json'))
data_meta.head()
```

```python
label_meta = pd.read_csv(os.path.join(dataset_root_path, 'reefset_labels_by_dataset.csv'), header=0)
label_meta.head()
```

```python
output_path = './data/ReefSet'
from lib.utils import make_unless_exits
make_unless_exits(output_path)
train_csv_file = os.path.join(output_path, 'train_annotations.csv')
test_csv_file = os.path.join(output_path, 'test_annotations.csv')

for label in label_meta['Class label']:
    sub_data = data_meta[data_meta['label'] == label]
    print(f'{label} size is: {len(sub_data)}')
```

```python
from sklearn.model_selection import train_test_split

for idx, label in enumerate(label_meta['Class label']):
    sub_data = data_meta[data_meta['label'] == label]
    train_set, test_set = train_test_split(sub_data, test_size=.3, random_state=123456)
    if idx == 0:
        ttl_train_set = train_set
        ttl_test_set = test_set
    else:
        ttl_train_set = pd.concat([ttl_train_set, train_set], ignore_index=True, axis=0)
        ttl_test_set = pd.concat([ttl_test_set, test_set], ignore_index=True, axis=0)
print(f'ttl_train_set size is: {len(ttl_train_set)}')
print(f'ttl_test_set size is: {len(ttl_test_set)}')
ttl_train_set.head()
```

```python
ttl_train_set.to_csv(train_csv_file)
ttl_test_set.to_csv(test_csv_file)
```
